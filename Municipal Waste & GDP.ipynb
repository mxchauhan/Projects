```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
waste = pd.read_csv("Total Municipal Waste Collected.csv")
```

```python
# Remove footnotes
waste = waste.iloc[:1587,:]
waste.columns = ["Country", "Year", "Value", "Value Footnotes", "Unit"]
waste["Value"] = waste["Value"].astype(float)
waste["Year"] = waste["Year"].astype(int)
```

```python
waste
```

```python
country_mean_waste = waste.groupby("Country")["Value"].aggregate(np.mean)
```

```python
country_mean_waste
```

```python
max_value = waste.loc[waste['Value'].idxmax()]
```

```python
max_value
```

```python
USA = waste.loc[lambda waste: waste['Country']=="United States of America", :]
```

```python
USA = USA.sort_values('Year')
## USA = USA.reset_index(drop=True)
USA = USA.set_index(USA["Year"])
```

```python
USA
```

```python
plt.figure(figsize=(8, 5))
plt.plot(USA["Value"])
plt.ylabel("Municipal Waste (1000 tonnes)")
plt.xlabel("Year")
plt.title("Total Municipal Waste Collected in US (1990-2014)")
plt.xticks(rotation = 0)
plt.locator_params(axis = 'x', nbins=6)
plt.show()
```

The visualization above represents the total municipal waste collected in the United States over the years of 1990 to 2014.  We can see a general increase overtime in the municipal waste that is collected, with a few drops.  The drop around 2008 may be due to the market crash that occured this year.  This general upward trend in waste collected may be due to an increase in consumption of resources over time.

```python
GDP = pd.read_csv("GDP Per Capita.csv")
GDP = GDP.iloc[:6194,:]
GDP.columns = ["Country", "Year", "Value", "Value Footnotes"]
GDP["Value"] = GDP["Value"].astype(float)
GDP["Year"] = GDP["Year"].astype(int)
```

```python
GDP
```

```python
USA_GDP = GDP.loc[lambda GDP: np.logical_and(GDP["Country"]=="United States",(GDP["Year"] < 2015))]
```

```python
USA_GDP = USA_GDP.sort_values("Year")
USA_GDP = USA_GDP.set_index(USA_GDP["Year"])
```

```python
USA_GDP
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x = USA_GDP["Value"], y = USA["Value"], color = "purple")
plt.xlabel("GDP Per Capita (PPP)")
plt.ylabel("Municipal Waste (1000 tonnes)")
plt.title("US GDP Per Capita vs. Municipal Waste (1990-2014)")
plt.show()
```

The visualizaiton above represents the relationship between GDP per capita and municipal waste collected over the years of 1990-2014.  Generally, as GDP increases municipal waste also increases.  This does not necessarily mean the GDP causes an increase in municipal waste, but the plot shows that there is a fairly strong postive correlation between the two variables.

```python
country_mean = pd.DataFrame({'Country':country_mean_waste.index, 'Value':country_mean_waste.values})
```

```python
country_mean
```

```python
country_mean = country_mean.loc[country_mean['Country'].isin(['United States of America',
                                               'China',
                                               'Japan',
                                               'Brazil',
                                               'Germany',
                                               'Iceland',
                                               'United Kingdom of Great Britain and Northern Ireland',
                                               'Australia',
                                               'Egypt'
                                              ])]
country_mean = country_mean.sort_values('Value', ascending=False)
country_mean = country_mean.set_index(country_mean['Country'])
country_mean.rename(index={'United Kingdom of Great Britain and Northern Ireland': 'UK', 'United States of America': 'USA'}, inplace=True)
country_mean
```

```python
ax = country_mean['Value'].plot(kind='bar', figsize = (10, 6))
plt.xlabel('Country')
plt.ylabel('Municipal Waste (1000 tonnes)')
plt.title('Average Municipal Waste Collected in Various Developed/ing Countries (1990-2014)')
plt.xticks(rotation = 45)
plt.show()
```

The visualization above represents the average municpial waste collected in various developed and developing countries over the years 1990-2014.  The countries shown the plot were chosen based on how developed they are and their population.  It is likely that countries with a greater population as well as the extent to which they are developed may impact the amount of municipal waste that is collected.  It is interesting to note that despite China being the most populated country in the world, the U.S. collects more municipal waste.  Unfortuantely other countries which would have been respectable comparisions, such as Russia, Canada, India, etc., were not included in the dataset.

```python
China_GDP = GDP.loc[lambda GDP: np.logical_and(GDP["Country"]=="China",(GDP["Year"] < 2015))]
```

```python
China_GDP = China_GDP.sort_values("Year")
China_GDP = China_GDP.set_index(China_GDP["Year"])
China_GDP
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1,2,1)
plt.plot(USA_GDP['Value'])
plt.ylabel('GDP Per Capita (PPP)')
plt.xlabel('Year')
plt.title('USA GDP Per Capita (1990-2014)')

plt.subplot(1,2,2)
plt.plot(China_GDP['Value'], color='red')
plt.ylabel('GDP Per Capita (PPP)')
plt.xlabel('Year')
plt.title('China GDP Per Capita (1990-2014)')

plt.tight_layout()
plt.show()
```

The visualization above represnets the GDP of the United States compared to the GDP of China over the years of 1990-2014.  The varying ranges of GDP should be noted as the United States' ranges from 25000-55000, while China's ranges from 2000-14000.  Both countries show an increase in GDP over time.  However, we can see the China's GDP is increasing at an increasing rate, while the GDP of the United States is increasing, but at a slightly decreasing rate.
